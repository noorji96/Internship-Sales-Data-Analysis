# To Reduce the noise in data , grouped the items with values less than 10 as "others"

```python
def handle_items(value):
    if(value in items_lessthan10):
        return 'others'
    else:
        return value
        
data['Item'] = data['Item'].apply( handle_items)
data['Item'].value_counts()
```

```python
Monthly_Sales =data.groupby('Month').sum().reset_index()
sns.catplot(y = 'Sales Amount', x = 'Month', data = Monthly_Sales, palette='Blues',kind="bar")
plt.xlabel('Year')
plt.ylabel('Sales Amount')
plt.title('Monthly Sales')
Monthly_Sales[['Month', 'Sales Amount']]
```
